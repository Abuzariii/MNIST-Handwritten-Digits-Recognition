```python
import tensorflow as tf
import numpy as np
```

```python
from tensorflow.keras.models import load_model

model = load_model("mnist_digit_recognition_model.h5")
```

```python
from tensorflow.keras.datasets import mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_test = x_test.reshape(10000, 28, 28, 1)
```

```python
y_pred = np.argmax(model.predict(x_test), axis=-1)
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_pred))
```

CleanShot 2020-11-30 at 10.46.45.png

```python
# Per-class accuracy
class_accuracy = 100 * confusion_matrix(y_test, y_pred).diagonal() / confusion_matrix(y_test, y_pred).sum(1)

for (i,ca) in enumerate(class_accuracy):
    print(f'Accuracy for {i} : {ca:.3f}%')
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
